## 1.3 Outlier Detection and Treatment

```python
def cap_outliers(df, column):
    """Caps outliers in a specified column using the IQR method."""
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)
    return df


# List of columns with outliers that need capping
columns_to_cap = ['T1', 'T2', 'T3', 'T4', 'T5', 'T6', 'T7', 'T8', 'RH_1','RH_2', 'RH_3', 'RH_5', 'RH_7', 'RH_8', 'RH_9','T_out', 'Press_mm_hg', 'RH_out', 'Windspeed','Visibility','Tdewpoint']

# Apply capping to each specified column
for column in columns_to_cap:
    data = cap_outliers(data, column)
```

#### Apply capping to features (independent variables) to handle outliers.
#### Do not cap or modify the target column (dependent variable) since it represents actual values you want to predict.

## 1.4 Data Scaling and Normalization

```python
# scaling using stadardization
from sklearn.preprocessing import StandardScaler
```

```python
# Define the features to scale (exclude 'date' and 'Appliances')
features_to_scale = [col for col in data.columns if col not in ['date', 'Appliances']]

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit and transform the selected features
data[features_to_scale] = scaler.fit_transform(data[features_to_scale])

# Display the scaled data
print(data.head())
```

## 1.5 Data Splitting

```python
# prompt: Data Splitting:

# Split data into features (X) and target (y)
X = data.drop(columns=['date', 'Appliances'])
y = data['Appliances']

# Split data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42) # 70% training and 30% test

# Print the shapes of the resulting datasets
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```
